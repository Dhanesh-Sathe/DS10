```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dataframe=pd.DataFrame(data=iris.data,columns=iris.feature_names)
```

```python
# features and their types
dataframe.info()
```

```python
dataframe.columns
```

```python
numeric_cols=dataframe.select_dtypes(include=['int','float64']).columns
```

```python
# histogram for each feature
for i in numeric_cols:
    sns.histplot(x=i,bins=8,kde=True,data=dataframe)
    plt.show()
```

```python
# box plot for each feature
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=col,data=dataframe)
    plt.show()
```

```python
# finding outliers using IQR
for col in numeric_cols:
    q1=dataframe[col].quantile(0.25)
    q3=dataframe[col].quantile(0.75)
    IQR=q3-q1
    upper_bound=q3+1.5*IQR
    lower_bound=q1-1.5*IQR
    outliers=dataframe[(dataframe[col]>upper_bound)| (dataframe[col]<lower_bound)]
    print(f"{col}--{len(outliers)}")
```
